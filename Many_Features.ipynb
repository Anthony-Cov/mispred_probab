```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from src.utils import Norm01, Nback, Metr, MovingAverage, seconds_to_str 
from src.features import NoiseFactor, RandWalk, CEmbDim, CHurst, KSEntr
from time import time


from os import listdir
```

```python
datadir='..\_DataSets\HugeStockMarket\Stocks'
dfiles=listdir(datadir)
len(dfiles)
#dfiles.index('ecf.us.txt')
```

```python
cols=['series','noise', 'walk','edim', 'hurst', 'ksent']
manyfeatures=pd.read_csv('mft.csv') # pd.DataFrame(columns=cols) #
strt=time()
for i, f in enumerate(dfiles[6000:]):
    df=pd.read_csv(datadir+'\\'+f)
    if len(df)<730:
        continue
    else:
        res=pd.DataFrame(columns=cols)
        df.drop(df[df.index<len(df)-730].index, inplace=True)
        df.reset_index(drop=True, inplace=True)
        ser=df.Close.values
        res['series']=[f[:-7]]
        res['noise']=[NoiseFactor(ser)]
        res['walk']=[RandWalk(ser)]
        res['edim']=[CEmbDim(ser)]
        res['hurst']=[CHurst(ser)]
        res['ksent']=[KSEntr(ser)]
        manyfeatures=pd.concat([manyfeatures, res], axis=0, ignore_index=True)
        if i%100==0:
            manyfeatures.to_csv('mft.csv', index=False)
            print(i, seconds_to_str(time()-strt), 'mft.csv saved.')  
manyfeatures.to_csv('mft.csv', index=False)
manyfeatures
```
